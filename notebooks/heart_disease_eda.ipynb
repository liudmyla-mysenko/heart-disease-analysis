## Step 10: Identify Categorical Columns

Categorical columns contain values that represent categories rather than numerical measurements.  
Identifying these columns is important because they often require encoding before machine learning models can use them.

In the Heart Disease dataset, some columns are stored as integers but actually represent categories.  
We need to identify which columns should be treated as categorical.

Examples of categorical features in this dataset include:

- **sex** — 0 = female, 1 = male  
- **cp** (chest pain type) — 4 categories  
- **fbs** (fasting blood sugar) — 0 or 1  
- **restecg** — 3 categories  
- **exang** — 0 or 1  
- **slope** — 3 categories  
- **ca** — number of major vessels (0–3)  
- **thal** — 3 categories (3, 6, 7)

We will identify these columns programmatically to prepare them for encoding in later steps.


````python
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

# Also include integer columns that represent categories
categorical_cols += ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

categorical_cols
````

## Step 11: Apply Label Encoding to One Binary Categorical Column

Label Encoding is used to convert categorical values into numerical form.  
It is especially simple for **binary categorical columns**, where the values represent only two categories.

In the Heart Disease dataset, several columns are binary, such as:

- **sex** — 0 = female, 1 = male  
- **fbs** — fasting blood sugar (0 or 1)  
- **exang** — exercise-induced angina (0 or 1)

For this step, we will apply Label Encoding to one binary column — for example, **exang**.

Since binary columns already contain numerical values (0 and 1), Label Encoding will not change the values, but this step demonstrates the encoding process.


````python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['exang_encoded'] = le.fit_transform(df['exang'])

df[['exang', 'exang_encoded']].head()
````

## Step 12: Apply One-Hot Encoding to One Categorical Column with Multiple Categories

One-Hot Encoding is used to convert categorical variables with more than two categories into separate binary columns.  
Each new column represents one category, with values of 0 or 1.

In the Heart Disease dataset, several columns have multiple categories.  
A good example is **cp** (chest pain type), which has four categories:

- 0 — Typical angina  
- 1 — Atypical angina  
- 2 — Non-anginal pain  
- 3 — Asymptomatic  

We will apply One-Hot Encoding to the `cp` column to create new binary columns representing each chest pain type.


````python
# One-Hot Encode the 'cp' column
cp_encoded = pd.get_dummies(df['cp'], prefix='cp')

# Add encoded columns to the dataset
df = pd.concat([df, cp_encoded], axis=1)

df.head()
````

## Step 13: Summary Statistics for a Numerical Column

To better understand the distribution of numerical data, we can calculate basic summary statistics such as:

- **Maximum value** (`max()`)
- **Minimum value** (`min()`)

- **Median** (`median()`)

These metrics help us identify the range of values, central tendency, and potential outliers.

In this step, we will apply these functions to the **age** column.


````python
# Apply summary statistics to the 'age' column
age_max = df['age'].max()
age_min = df['age'].min()
age_mean = df['age'].mean()
age_median = df['age'].median()

age_max, age_min, age_mean, age_median
````

## Step 15: Filter the Dataset Using Two Conditions

Filtering allows us to extract specific subsets of data that meet certain criteria.  
This is useful for exploring patterns, identifying risk groups, or preparing data for modeling.

In this step, we will filter the dataset using **two conditions**.  
For example, we can select patients who:

- are older than 50  
- and have chest pain type equal to 2 (non-anginal pain)

This gives us a focused subset of patients with similar characteristics.


````python
filtered_df = df[(df['age'] > 50) & (df['cp'] == 2)]
filtered_df.head()
````
